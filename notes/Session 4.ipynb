###Fetch commit data

### Most common weekday and hour

###Commits per day of the week

```python
commitsByDay=commits.count(1).resample('D',how='sum').fillna(0)
```

```python
commitsByDay
```

```python
commitsByDay.groupby(commitsByDay.index.dayofweek).agg([np.sum, np.mean, np.std])
```

###Commits per hour of the day

```python
commitsByHour=commits.count(1).resample('H',how='sum').fillna(0)
```

```python
commitsByHour.groupby(commitsByHour.index.hour).agg([np.sum, np.mean, np.std])
```

### Commits per hour per day

```python
grouped = commitsByHour.groupby(commitsByHour.index.dayofweek)

for key,group in grouped:
    print key
    print group.groupby(group.index.hour).agg([np.sum, np.mean, np.std])
    
```

###Most active users

```python
counts = commits.apply(lambda a: a.count())
counts.sort()
counts
```

###Who comitted at 2 AM

```python
import pandas
commits.applymap(lambda a : 0 if pandas.isnull(a) else 1).resample('H',how='sum').fillna(0).groupby(commitsByHour.index.hour).sum().ix[2]
```

### Average commit message length

```python
averageCommitLen = commits.applymap(lambda a : np.NAN if pandas.isnull(a) else len(a)).mean()
averageCommitLen.sort()
averageCommitLen
```
